### Trap 1 — predict, then run

```python
def enroll(name, roster=[]):
    roster.append(name)
    return roster
enroll('Ana')
enroll('Ben')
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `['Ana', 'Ben']`

A default value is created ONCE, at def time, so the same list persists across calls. Use `roster=None` then `roster = roster or []`.

</details>

### Trap 2 — predict, then run

```python
def show_score(s):
    print(s)
show_score(91) is None
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

Printing is not returning. A function with no `return` gives `None`.

</details>

### Trap 3 — predict, then run

```python
makers = [lambda: score for score in [88, 91, 73]]
[make() for make in makers]
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `[73, 73, 73]`

Closures capture the VARIABLE `score`, not its value; by call time the loop has left `score = 73`. Fix by binding it: `lambda score=score: score`.

</details>

### Trap 4 — predict, then run

```python
def grade(n):
    return grade(n - 1)   # forgot the base case
grade(3)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `RecursionError: maximum recursion depth exceeded`

Every recursive function needs a base case. Without one Python stops at its recursion limit (~1000 deep) with RecursionError.

</details>
