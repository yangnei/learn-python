### Trap 1 — predict, then run

```python
ana_scores = [88, 91]
ben_scores = [88, 91]
ana_scores is ben_scores
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`==` compares VALUES (and `ana_scores == ben_scores` is True); `is` compares IDENTITY — whether they are the same object in memory. Ana's and Ben's lists just happen to match. Use `==` for values and keep `is` for `None`.

</details>

### Trap 2 — predict, then run

```python
roster = ['Ana', 'Ben']
backup = roster
roster.append('Cara')
backup
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `['Ana', 'Ben', 'Cara']`

`backup = roster` does not copy: both names point at one list, so appending through `roster` shows up in `backup` too. Copy with `list(roster)` or `roster[:]`.

</details>

### Trap 3 — predict, then run

```python
attended = True
attended == 1
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

`bool` is a subclass of `int`: `True == 1` and `False == 0`, which is why a spreadsheet's 1/0 lines up with Python's True/False.

</details>

### Trap 4 — predict, then run

```python
score = 5
bonus = True   # earned a bonus point?
score + bonus
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `6`

`True` acts as `1` (and `False` as `0`) in arithmetic — a quick way to add a bonus point for a flag.

</details>

### Trap 5 — predict, then run

```python
passed = [True, False, True, True]   # did each student pass?
sum(passed)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `3`

Summing booleans counts the `True`s — a handy way to count how many students passed.

</details>

### Trap 6 — predict, then run

```python
homework = 0.1
exam = 0.2   # their weights in the final grade
homework + exam
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `0.30000000000000004`

Floats are stored in binary, where 0.1 and 0.2 have no exact representation, so the tiny errors add up.

</details>

### Trap 7 — predict, then run

```python
weight = 0.1 + 0.2   # homework + exam
weight == 0.3
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Because `0.1 + 0.2` is `0.30000000000000004`. Never test computed scores with `==`; use `math.isclose(a, b)`.

</details>

### Trap 8 — predict, then run

```python
missing_score = float('nan')
missing_score == missing_score
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`NaN` (a missing/invalid score) is defined to equal nothing, not even itself. Test it with `math.isnan(x)`.

</details>

### Trap 9 — predict, then run

```python
total = 7   # two students' scores added up
total / 2
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `3.5`

`/` is always float (true) division, so the mean comes out as `3.5`. Use `//` when you want a whole number.

</details>

### Trap 10 — predict, then run

```python
penalty = -7   # points to spread over 2 assignments
penalty // 2
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `-4`

`//` floors toward negative infinity, not toward zero, so `-7 // 2 == -4` (it deducts 4 from each, not 3).

</details>

### Trap 11 — predict, then run

```python
score = '5'   # read from a CSV
score == 5
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Python does no automatic text/number conversion, so a CSV string and a number are simply unequal (no error). Convert first: `int(score) == 5`.

</details>

### Trap 12 — predict, then run

```python
score = '5'
5 > score
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `TypeError: '>' not supported between instances of 'int' and 'str'`

Ordering a number against text raises TypeError — there is no sensible order. Convert first: `5 > int(score)`.

</details>

### Trap 13 — predict, then run

```python
scores_list = [88, 91]
scores_tuple = (88, 91)
scores_list == scores_tuple
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

A list never equals a tuple, whatever the contents.

</details>

### Trap 14 — predict, then run

```python
passed = True
type(passed) is int
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

`type()` is exact, and `passed`'s type is `bool`, not `int`. Use `isinstance` when you want subclass-aware checks.

</details>

### Trap 15 — predict, then run

```python
passed = True
isinstance(passed, int)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

`isinstance` respects subclassing, and `bool` IS a kind of `int`.

</details>

### Trap 16 — predict, then run

```python
answer = '0'   # what a student typed on the survey
bool(answer)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `True`

Any NON-EMPTY string is truthy, including '0' and 'False'. Only the empty string is falsy — so convert survey text before testing it.

</details>

### Trap 17 — predict, then run

```python
submissions = []
bool(submissions)
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

Empty containers (`[]`, `{}`, `''`, `0`, `None`) are all falsy, which is why `if submissions:` reads as 'are there any?'.

</details>

### Trap 18 — predict, then run

```python
id_from_csv = int('257')
id_from_form = int('257')   # the same student ID, parsed twice
id_from_csv is id_from_form
```

<details>
<summary><strong>What really happens — and why</strong></summary>

Result: `False`

CPython pre-caches small ints (-5..256), so `is` accidentally looks True there; 257 built at runtime are separate objects. Compare values with `==`, never `is`.

</details>
